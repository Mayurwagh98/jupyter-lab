```python
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]

oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.plot(years, oscar_revenue)
plt.title("Oscar movies revenue")
plt.xlabel("Years")
plt.ylabel("Revenue in $M")
```

```python
# non oscar vs oscar
# Non-Oscar Winner Movies

non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]

non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(years, oscar_revenue, "o--y",label="Oscar")
plt.plot(years, non_oscar_revenue, marker="<",color="red",linestyle="--",label="Non-Oscar")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.title("Oscar vs. Non-Oscar movies revenue")
plt.legend()
```

```python
# multi bar chart
non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

x = np.arange(len(years))
width= 0.4

plt.bar(x-width/2, oscar_revenue, width, label="Oscar")
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar")
plt.xlabel("Years")
plt.ylabel("Revenue")
plt.title("Oscar vs. Non-Oscar")
plt.xticks(x, years)

for i in range(len(years)):
    plt.text(x[i]-width/2, oscar_revenue[i] + 10, str(oscar_revenue[i]), ha="center")
    
for j in range(len(years)):
    plt.text(x[j]+width/2, non_oscar_revenue[j] + 10, str(non_oscar_revenue[j]), ha="center")
```

```python
# scatter plot
people = [ "Person A", "Person B", "Person C", "Person D", "Person E", "Person F", 
           "Person G", "Person H", "Person I", "Person J"];

age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65];
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150];

plt.scatter(age, bp, s=bp, cmap="OrRd", c=bp)
plt.grid()
plt.title("Blood pressure based on age")
plt.xlabel("Age")
plt.ylabel("BP")
plt.colorbar(label="BP")

for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i], bp[i]), xytext=(age[i], bp[i] + 1))
    
plt.xlim(min(age), max(age)+10)
plt.ylim(min(bp), max(bp)+10)
```

```python
# multiplot scatter
cities = ["City A", "City B", "City C", "City D", "City E"]

# Winter: Temperature (°C) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer: Temperature (°C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]

plt.scatter(winter_temp, winter_humidity,label="Winter")
plt.scatter(summer_temp, summer_humidity, label="Summer")
plt.xlabel("Temp")
plt.ylabel("Humidity")
plt.legend()
plt.grid()

for i in range(len(winter_temp)):
    plt.annotate(winter_temp[i], xy=(winter_temp[i], winter_humidity[i]), xytext=(winter_temp[i], winter_humidity[i]+2))

for j in range(len(summer_temp)):
    plt.annotate(summer_temp[j], xy=(summer_temp[j], summer_humidity[j]), xytext=(summer_temp[j], summer_humidity[j]+2))

plt.ylim(min(winter_temp)+7, max(winter_humidity)+7)
```

```python
# pie plot
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

explode=[0,0,0,0.1,0]

plt.pie(amounts, autopct="%1.1f%%",explode=explode, wedgeprops={
    "edgecolor":"blue",
    "linewidth":1,
    "linestyle":"--"
},
       startangle=90,
       shadow=True)
plt.xlabel("Expenses")
plt.ylabel("Amounts")
```
