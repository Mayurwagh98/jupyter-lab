```python
import requests
```

```python
URL = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(URL)
```

```python
if res.status_code == 200:
    print(res.headers)

with open("scrapped_data.html", "w") as file:
    file.write(res.text)
```

```python
# Beautifulsoup

from bs4 import BeautifulSoup

with open("scrapped_data.html","r") as file:
    html_content = file.read()

soup = BeautifulSoup(html_content,"lxml")
```

```python
all_h3 = soup.find_all("h3")

all_countries = []

for h3 in all_h3:
    name = h3.get_text(strip = True)
    population = h3.find_next("div").select("span.country-population")[0].get_text(strip = True)
    all_countries.append([name, population])
```

```python
import pandas as pd
df = pd.DataFrame(all_countries, columns=["Country","Population"])
df
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
