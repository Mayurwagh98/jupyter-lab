```python
# Q2. Start by loading the TITANIC dataset.
# A. Display only columns:
# Name
# Sex
# Age
# Fare
# Survived

# B. Select passengers who:
# Are female
# Fare > 30

# C. Group by Pclass and compute:
# Survival rate
# Average Fare
# Average Age
```

```python
import pandas as pd
```

```python
# A. Display only columns:
# Name
# Sex
# Age
# Fare
# Survived
df = pd.read_csv("Titanic-Dataset.csv")
df
df[["Name", "Sex", "Age", "Fare", "Survived"]]
```

```python
# B. Select passengers who:
# Are female
# Fare > 30
# df[(df["Sex"] == "female") & (df["Fare"] > 30)]
df.query("Sex == 'female' & Fare > 30")[["Sex", "Fare"]]
```

```python
# C. Group by Pclass and compute:
# Survival rate
# Average Fare
# Average Age

df.groupby("Pclass").agg(
    survival_rate = ("Survived","mean"),
    avg_fare = ("Fare","mean"),
    avg_age = ("Age","mean")
)
```
