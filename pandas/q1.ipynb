```python
# Q1. Start by loading the IRIS dataset.

# A. Display the following information:
# First 10 rows
# Shape & Data types
# Summary statistics (mean, std, min, max)

# B. Select only those rows where:
# petal_length > 4.5
# species = "Iris-virginica"

# C. Group by species and compute:
# Average sepal_length
# Maximum petal_width
# Standard deviation of sepal_width

# D. Create a new column:
# petal_ratio = petal_length / petal_width
# For each species, find the average petal_ratio.
```

```python
import pandas as pd
```

```python
# A. Display the following information:
# First 10 rows
# Shape & Data types
# Summary statistics (mean, std, min, max)

df = pd.read_csv("pandas/IRIS.csv")
df
df.head(10)
df.shape
df.dtypes
df.describe()
```

```python
# B. Select only those rows where:
# petal_length > 4.5
# species = "Iris-virginica"
df[(df["petal_length"] > 4.5) & (df["species"] == "Iris-virginica")]
```

```python
# C. Group by species and compute:
# Average sepal_length
# Maximum petal_width
# Standard deviation of sepal_width

df.groupby("species")["sepal_length"].mean()
df.groupby("species")["petal_width"].max()
df.groupby("species")["sepal_width"].std()

# better version
df.groupby("species").agg({
    "sepal_length":"mean",
    "petal_width":"max",
    "sepal_width":"std"
})
# or
df.groupby("species").agg(
    avg_sepal_length=("sepal_length","mean"),
    max_petal_width=("petal_width","max"),
    std_deviation=("sepal_width","std")
)
```

```python
# D. Create a new column:
# petal_ratio = petal_length / petal_width
# For each species, find the average petal_ratio.
df2 = df.copy()
# new column
df2["petal_ratio"] = df2["petal_length"] / df2["petal_width"]
# avg for each species
df2.groupby("species")["petal_ratio"].mean()
```
