```python
import json
```

```python
def load_data(filename):
    with open(filename, "r") as file:
        data = json.load(file)
    return data    
```

```python
data = load_data("store_data.json")
print(data)
```

 <!-- get meaningful insights from the data -->
<!-- 1. what's avg rating? -->
<!--- 2. What % of users gave poor rating (<3)? -->

```python
# cleaning the data
def clean_data(data):
    # consistency in rating
    text_to_num = {"one":1, "two":2, "three":3, "four":4, "five":5}
    cleaned_users = []
    unique_users = set()
    
    for user in data:
        raw_rating = str(user["rating"]).strip().lower()
        if (raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        # adding age if missing
        raw_age = user.get('age')
        if(raw_age == None):
            user['age'] = None

        # removing duplcate data
        if(user['name'].strip() in unique_users):
            continue
        cleaned_users.append(user)
        unique_users.add(user['name'])

    return cleaned_users
```

```python
new_data = clean_data(data)
```

```python
# get meaningful insights from the data
# 1. what's avg rating?
# 2. What % of users gave poor rating (<3)?
def get_insights(data):
    total_avg_rating = 0
    poor_rating = 0

    for user in data:
        total_avg_rating += float(user['rating'])

        if(float(user['rating']) < 3):
            poor_rating += 1
            
    # average rating
    average_rating = total_avg_rating/len(data)
    print(f"avg rating: {average_rating:.2f}")

    # poor rating
    total_rating = (poor_rating /len(data) * 100)
    print(f"% of poor rating: {total_rating:.2f}")

get_insights(new_data)
```

```python
# product recommendation
# 1. if user rating >= 4, recommend same brand products
# 2. if user rating < 4, recommend different brand products

def product_recommend(data):
    recommendations = []
    for user in data:
        current_recommendation = {}
        current_recommendation["name"] = user["name"]
        if float(user["rating"]) >= 4:
            current_recommendation["bran"] = "Apple"
        else:
            current_recommendation["brand"] = "Samsung"
        recommendations.append(current_recommendation)
    return recommendations
product_recommend(new_data)
```
