```python
# Q2. For the Seaborn penguins dataset, do the following:
# Plot flipper_length_mm vs body_mass_g
# Color by species
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
penguins[["flipper_length_mm","body_mass_g","species"]]
```

```python
sns.scatterplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species"
)
plt.title("Flipper vs. Body mass")
plt.xlabel("Flipper length")
plt.ylabel("Body mass")
```

```python
# Q3. For the Seaborn tips dataset, do the following:
# Create a box plot of total_bill for each day
# Add hue="sex"
# Identify which day has the highest median bill
# Comment on outliers

tips = sns.load_dataset("tips")

sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)
plt.title("Total bill for each day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.tight_layout()
```

```python
# Q4. For the Seaborn tips dataset, do the following:
# Create a scatter plot of total_bill vs tip
# Color points by sex
# Use different markers for smoker
# Add a title and axis labels
# Analyze: Do smokers tend to tip differently?

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
)
plt.title("Total bill vs. Tip")
plt.xlabel("Total bill")
plt.ylabel("Tip")
plt.tight_layout()
```

```python
# Q5. For the Seaborn tips dataset, do the following:
# Create a pivot table of average tip
# Rows: day
# Columns: time
# Plot the pivot table as a heatmap
# Annotate values
# Interpret which day–time combination performs best

tips_pivot = tips.pivot_table( values="tip",
    index="day",
    columns="time",
    aggfunc="mean")

sns.heatmap(
    tips_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)
```
